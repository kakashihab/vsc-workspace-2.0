```python
import pandas as pd
```

```python
from matplotlib import pyplot
```

```python
x = [1,2,3,4,5]
y = [10,5,8,16,23]
z = [1,4,3,6,5]
pyplot.plot(x,y)
pyplot.plot(x,z)
pyplot.legend(["y values", "z values"])
pyplot.xlabel("Sales")
pyplot.ylabel("Month")
pyplot.title("My sales graph")
pyplot.show()
```
